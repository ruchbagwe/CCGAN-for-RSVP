## Preproccessing and Saving Data
### Function to Convert a Directory of Images to 64x64x3 Numpy Arrays

```python
def convert_dir_to_imgs(img_dir):
     
    # get file names
    nm_imgs = os.listdir(img_dir)
  
    #convert each file into a numpy array
    img_shape=(64,64,3)
    images = []
    for i, myid in enumerate(nm_imgs):
        image = load_img(img_dir + "/" + myid,
                     target_size=img_shape[:2])
        image = img_to_array(image)/255
        #image = (img_to_array(image)- 127.5)/127.5
        images.append(image)
    images = np.array(images)
  
  
    return images
```

## Creating the Unshuffled Dataset

```python
import numpy as np

tot_num_images=29710
num_classes=20
eeg_embedding_dim=126

dataset = [np.zeros((tot_num_images,num_classes)),np.zeros((tot_num_images,eeg_embedding_dim)),np.zeros((tot_num_images,64,64,3))]

ix = 0
cum_total = 0

for img_class in np.arange(20):
    # get number of images
    num_images = img_all[img_class].shape[0]
    cum_total = cum_total + num_images
    # fill in dataset with the images along with num_images amount of embeddings and labels
    for img_ix in np.arange(num_images):
        #fill in labels
        dataset[0][ix] = label_all_np[img_class] 
        # fill in eeg embeddings
        dataset[1][ix] = eeg_emb_all_np[img_class]
        # fill in images
        dataset[2][ix] = img_all_np[img_class][img_ix]
        ix=ix+1
```

## Creating and Saving the Shuffled Dataset

### Shuffling

```python
shuffle_indices = np.arange(tot_num_images)
np.random.seed(10)
np.random.shuffle(shuffle_indices)
shuffle_indices
```

### Creating

```python
d0 = dataset[0][shuffle_indices]
d1 = dataset[1][shuffle_indices]
d2 = dataset[2][shuffle_indices]
dataset_shuf = [d0,d1,d2]
```
